```python
#this is my second version of gradient descent

import time
def run_gradient_descent(num_iter,alpha,w,b,x,y):

  w_history=[]
  j_history=[]


  for i in range(num_iter):
    dj_dw,dj_db=compute_gradient(w,b,x,y)
    w-= alpha*dj_dw
    b-=alpha*dj_db
    w_history.append(w)
    cost=compute_cost(w,b)
    j_history.append(cost)

    if i%100==0 or i==num_iter-1:
     print(f"For iteration {i}, the cost is {cost:.3f}, value of w is {w:.3f}, value of b is {b:.3f},dj_dw is {dj_dw}, dj_db is {dj_db}")  #.3f for showing till only 3 decimal places
     time.sleep(1)
  return w,b,j_history,w_history
```
