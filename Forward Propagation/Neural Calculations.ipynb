## **Neural Functions**

```python
import numpy as np
```

```python
def sigmoid(z):
  g = 1/(1+np.exp(-z))
  return g
```

```python
def relu(z):
  return np.maximum(0,z)
```

```python
def layer(w,x,b):
  a = w@x + b #w.shape = (n,m) n = no of neurons, m = no of features, x.shape = (l,m) l = no of datapoints, b = column vector (n,1)
  return a #a.shape = (n,l)
```

```python
def sequencial(W1,X,B1,W2,B2):  #for neural network of 2 layers
  a1 = layer(W1,X,B1)
  a2 = layer(W2,a1,B2) # W2.T@a1.T + B2 , W2.T.shape = (1,l) B2 is a scalar a1.T.shape = l,n
  return a2
```

```python
w1 = np.array([
    [0.1,0.2,0.3],
    [0.5,0.3,0.1] #3 features and 2 neurons, shape = 2,3
])
# 2,3 @ 3,4 = 2,4 = activations because 2 neurons, so actiavtion is of format [a1]
#                                                                             [a2]
x = np.array([
    [3,5,4],
    [7,6,3],
    [3,5,6],
    [4,3,1]   #3 features and 4 data points, shape = 4,3
])

b1 = np.array([
    [0.4],
    [0.9],
     #1 bias for each neuron of layer 1
])
#  w2@a2 + b2... we want of shape 4,1 or 1,4 basically 1 for each data point....no
#  4,1 @ 2,4 ...no

w2 = np.array([  #shape = 1,2
    [0.2,0.1] #note: we need one weight for each output avctivation of each neuron of prev layer basically 2 weights to multiply to a1 and a2 respectively
])

b2 = np.array([[0.9]])
```

```python
sequencial(w1,x.T,b1,w2,b2)
```
