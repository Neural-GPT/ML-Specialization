```python
import numpy as np
```

```python
''' return dj_dw  and dj_db
J(w,b)X = 1/2m * ∑(pred[i]-y[i])^2
dj_dw = 1/m * ∑(pred[i]-y[i])*x[i] --> after derivative of x^2 because it is mean squared function, it will be 2*x and 2 will be terminated by 1/2m
dj_db = 1/m * ∑(pred[i]-y[i]) --> b is a constant '''
x=np.array([5,4,7,8,2])
w=3
b=7
y=np.array([25,20,32,37,19])

def compute_gradient(w,b,x,y):
  m=len(x)
  dj_dw=0
  dj_db=0
  pred=0
  pred = w*x+b
  error= pred-y
  dj_dw= np.sum((error*x)/m)
  dj_db= np.sum((error)/m)

  return dj_dw, dj_db

```

```python
 dj_dw, dj_db=compute_gradient(w,b,x,y)
```
